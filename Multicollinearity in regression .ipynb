```python
def correlation(df,threshold):
    col_cor=set()
    corr_matrix=df.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colnmae=corr_matrix.columns[i]
                col_cor.add(colnmae)
    return col_cor
```
